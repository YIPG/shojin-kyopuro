## num2word
1000億単位まで対応しています。エラーはコード1で抜けるようになっています。
小数点以下もいけます.


```python
# 方針
# 英語の数字表記は3桁ずつ区切られている。
# まずエラー処理をかく。そうすればありえない数字には対応しなくて良い。

# たとえば102と02を比較する。このとき02はエラーだが、102はok. 先に02などのエラー（先頭に0がつく）を排除することで分岐を考えなくて良い。

# また数字は桁の小さいものから埋めていくことにする。たとえば10000と10では, Ten thousand, Tenと共通部分が多い。
# 3桁単位で考えていく.


import sys

target = input()

target = target.split(".")

ans_l = []

# 整数部分の計算

num = target[0]

def syosu2word(i):
    if i=="0":
        ans_l.append("zero")
    if i=="1":
        ans_l.append("one")
    if i=="2":
        ans_l.append("two")
    if i=='3':
        ans_l.append("three")
    if i=='4':
        ans_l.append("four")
    if i=='5':
        ans_l.append("five")
    if i=='6':
        ans_l.append("six")
    if i=='7':
        ans_l.append("seven")
    if i=='8':
        ans_l.append("eight")
    if i=='9':
        ans_l.append("nine")

def num2word(i):
    if i=="1":
        ans_l.append("one")
    if i=="2":
        ans_l.append("two")
    if i=='3':
        ans_l.append("three")
    if i=='4':
        ans_l.append("four")
    if i=='5':
        ans_l.append("five")
    if i=='6':
        ans_l.append("six")
    if i=='7':
        ans_l.append("seven")
    if i=='8':
        ans_l.append("eight")
    if i=='9':
        ans_l.append("nine")
        
def hutaketa(i):
    if i=="10":
        ans_l.append("ten")
    if i=="11":
        ans_l.append("eleven")
    if i=="12":
        ans_l.append("twelve")
    if i=='13':
        ans_l.append("thirteen")
    if i=='14':
        ans_l.append("fourteen")
    if i=='15':
        ans_l.append("fifteen")
    if i=='16':
        ans_l.append("sixteen")
    if i=='17':
        ans_l.append("seventeen")
    if i=='18':
        ans_l.append("eighteen")
    if i=='19':
        ans_l.append("nineteen")

def twenty_toka(i):
    if i=="2":
        ans_l.append("twenty")
    if i=='3':
        ans_l.append("thirty")
    if i=='4':
        ans_l.append("fourty")
    if i=='5':
        ans_l.append("fifty")
    if i=='6':
        ans_l.append("sixty")
    if i=='7':
        ans_l.append("seventy")
    if i=='8':
        ans_l.append("eighty")
    if i=='9':
        ans_l.append("ninety")
    
# エラーは存在しない前提なので、001はone, 012はtwelve, 102はone hundred twoとなるようにする。
def sanketa2word(num):
    if(num[0] != "0"):
        num2word(num[0])
        ans_l.append("hundred")
    if(num[1] != "0"):
        if(num[1] == "1"):
            hutaketa(num[1]+num[2])
            return
        else:
            twenty_toka(num[1])
    if(num[2]!= "0"):
        num2word(num[2])

# 上記のsanketa2wordで再利用が可能。あとは細かい場合分け

# 0
if len(num) == 1 and num[0]=="0":
    ans_l.append("zero")

# 012はエラー
elif num[0] == "0":
    sys.exit(1)
    
# 文字もエラー
if num.isdecimal() is False:
    sys.exit(1)

else:
    if(1<= len(num) <=3):
        num = num.zfill(3)
        sanketa2word(num)
    if(4<= len(num) <= 6):
        num = num.zfill(6)
        sanketa2word(num[-6:-3])
        ans_l.append("thousand")
        sanketa2word(num[-3:])
    if(7 <= len(num) <= 9):
        num = num.zfill(9)
        sanketa2word(num[-9:-6])
        ans_l.append("million")
        sanketa2word(num[-6:-3])
        ans_l.append("thousand")
        sanketa2word(num[-3:])
    if(10 <= len(num) <= 12):
        num = num.zfill(12)
        sanketa2word(num[-12:-9])
        ans_l.append("billion")
        sanketa2word(num[-9:-6])
        ans_l.append("million")
        sanketa2word(num[-6:-3])
        ans_l.append("thousand")
        sanketa2word(num[-3:])
        
        
# 小数部分
if len(target) == 2 and target[1].isdecimal() is True:
    ans_l.append("point")
    for i in target[1]:
        syosu2word(i)

# 小数以下に数字以外の文字
elif len(target) >= 2:
    sys.exit(1)
        
ans_l[0] = ans_l[0].capitalize()
ans_str = " ".join(ans_l)
print(ans_str)
```
